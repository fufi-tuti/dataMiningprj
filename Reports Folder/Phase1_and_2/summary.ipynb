```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


from google.colab import files
uploaded = files.upload()

file_path = '/content/Student_Depression_Dataset.xlsx'
data = pd.read_excel(file_path)


sample_data = data.sample(n=50)
print("Sample Data:")
print(sample_data)


numeric_data = data.select_dtypes(include=[np.number])
mean_values = numeric_data.mean()
std_values = numeric_data.std()

print("\nMean Values:")
print(mean_values)
print("\nStandard Deviation Values:")
print(std_values)

plt.figure(figsize=(8, 5))
sns.histplot(data['Depression'], bins=20, kde=True, color='blue')
plt.title('Distribution of Depression Scores')
plt.xlabel('Depression Score')
plt.ylabel('Frequency')
plt.show()


plt.figure(figsize=(8, 5))
sns.scatterplot(x=data['Sleep Duration'], y=data['Depression'], alpha=0.6, edgecolor='black', s=50)
plt.title('Scatter Plot of Depression Score vs Sleep Hours')
plt.xlabel('Sleep Hours')
plt.ylabel('Depression Score')
plt.grid(True)
plt.show()
```
