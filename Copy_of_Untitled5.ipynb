```python
# From DataCleaning.ipynb
import pandas as pd
file_path = "/content/Student Depression Dataset.csv"

df = pd.read_csv(file_path, encoding='utf-8')
df.head()

missing_rows = df[df.isnull().any(axis=1)]
print("\n Row with missing vaue:")
print(missing_rows)

df.fillna(df.select_dtypes(include=['number']).mean(), inplace=True)
for column in df.select_dtypes(include=['object']).columns:
    if df[column].isnull().sum() > 0:
        df[column].fillna(df[column].mode()[0], inplace=True)

missing_rows = df[df.isnull().any(axis=1)]
print("\n After cleaning")
print(missing_rows)

import numpy as np
Q1 = df.select_dtypes(include=['number']).quantile(0.25)
Q3 = df.select_dtypes(include=['number']).quantile(0.75)
IQR = Q3 - Q1
outliers = ((df.select_dtypes(include=['number']) < (Q1 - 1.5 * IQR)) |
            (df.select_dtypes(include=['number']) > (Q3 + 1.5 * IQR)))
print("\n Number of outliers :")
print(outliers.sum())

df_cleaned = df[~((df.select_dtypes(include=['number']) < (Q1 - 1.5 * IQR)) |
                   (df.select_dtypes(include=['number']) > (Q3 + 1.5 * IQR))).any(axis=1)]
print(f"\nNumber of rows before cleaning: {df.shape[0]}")
print(f"Number of rows after cleaning: {df_cleaned.shape[0]}")

# From Distribution.ipynb (if any code cells—add here similarly)

# From studentsDepression2.ipynb (if any code cells—add here similarly)

# From summary.ipynb (if any code cells—add here similarly)

# Additional processing and uploads
from google.colab import files
uploaded = files.upload()

# Load the cleaned dataset from local file
file_path = "cleaned_dataset.csv"
df_cleaned = pd.read_csv(file_path)

# Display first few rows of the cleaned dataset
print("Cleaned Dataset:")
print(df_cleaned.head())

# ✅ Step 1: Cleaning Column Names
df_cleaned.columns = df_cleaned.columns.str.strip()

# ✅ Step 2: Defining Numerical Features
numerical_features = ['Age', 'Academic Pressure', 'Work Pressure', 'CGPA',
                      'Study Satisfaction', 'Job Satisfaction', 'Work/Study Hours']

# ✅ Step 3: Displaying the Transformed Data
print("✅ Normalization Complete! Here are the first 5 rows:")
print(df_cleaned.head())

# إعادة تنظيف الأعمدة مرتين كما في الكود الأصلي
df_cleaned.columns = df_cleaned.columns.str.strip()
df_cleaned.columns = df_cleaned.columns.str.replace("  ", " ")
print(df_cleaned.columns.tolist())

# ✅ Select only numerical columns once
numerical_df = df_cleaned.select_dtypes(include=['number'])

# ✅ Check for skewness
skew_values = numerical_df.skew()
print("Skewness of Numerical Columns:")
print(skew_values)

# ✅ Optional: Identify which columns need discretization FIRST
print("\nRecommended for Discretization (Skew > 1 or Skew < -1):")
discretization_candidates = skew_values[abs(skew_values) > 1]
print(discretization_candidates)

# ✅ Feature Scaling (Normalization)
from sklearn.preprocessing import MinMaxScaler, LabelEncoder, OneHotEncoder
scaler = MinMaxScaler()
df_cleaned[numerical_features] = scaler.fit_transform(df_cleaned[numerical_features])

# ✅ Encoding Categorical Features
binary_features = ['Gender', 'Have you ever had suicidal thoughts ?', 'Financial Stress',
                   'Family History of Mental Illness', 'Depression']
label_encoder = LabelEncoder()
for feature in binary_features:
    df_cleaned[feature] = label_encoder.fit_transform(df_cleaned[feature])

# ✅ Ordinal Encoding for Ordered Categories
sleep_duration_mapping = {'Less than 5 hours': 0, '5-6 hours': 1, '7-8 hours': 2, 'More than 8 hours': 3}
degree_mapping = {'Class 12': 0, 'B.Sc': 1, 'M.Tech': 2, 'Ph.D': 3}
df_cleaned['Sleep Duration'] = df_cleaned['Sleep Duration'].map(sleep_duration_mapping)
df_cleaned['Degree'] = df_cleaned['Degree'].map(degree_mapping)

# ✅ One-Hot Encoding for Cities and Dietary Habits
df_cleaned = pd.get_dummies(df_cleaned, columns=['City', 'Dietary Habits'])

# ✅ Save Transformed Dataset
df_cleaned.to_csv("final_transformed_dataset.csv", index=False)
print("✅ Final Data Processing Complete! Preview:")
print(df_cleaned.head())

df_cleaned.to_csv("Processed_Student_Depression_Dataset.csv", index=False)
print("✅ Processed dataset saved successfully!")

# Visualization from summary.ipynb
import matplotlib.pyplot as plt
import seaborn as sns

df_excel = pd.read_excel('/content/Student_Depression_Dataset.xlsx')
df_sample = df_excel.sample(n=100)

plt.figure(figsize=(8, 5))
sns.countplot(x=df_sample["Gender"], palette="pastel")
plt.title("Distribution of Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(8, 5))
sns.countplot(x=df_sample["Profession"], palette="pastel")
plt.title("Distribution of Profession")
plt.xlabel("Profession")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(5,3))
df_sample["Age"].plot.hist(bins=20, edgecolor="black")
plt.title("Histogram of Age")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(5,3))
df_sample["CGPA"].plot.hist(bins=20, edgecolor="black")
plt.title("Histogram of CGPA")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(5,3))
df_sample["Financial Stress"].plot.hist(bins=20, edgecolor="black")
plt.title("Histogram of Financial Stress")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(5,3))
df_sample["Academic Pressure"].plot.hist(bins=20, edgecolor="black")
plt.title("Histogram of Academic Pressure")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(5,3))
df_sample["Study Satisfaction"].plot.hist(bins=20, edgecolor="black")
plt.title("Histogram of Study Satisfaction")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(5,3))
df_sample["Work/Study Hours"].plot.hist(bins=20, edgecolor="black")
plt.title("Histogram of Work/Study Hours")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(x='CGPA', y='Sleep Duration', hue=df_sample['Depression'], data=df_sample, palette="bright")
plt.title("Sleep Duration and CGPA")
plt.xlabel("CGPA")
plt.ylabel("Sleep Duration")
plt.show()

class_counts = df_sample['Depression'].value_counts()
plt.figure(figsize=(6, 4))
class_counts.plot(kind="bar")
plt.title("Class Imbalance Analysis")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()

# Dataset overview
df_info = pd.read_excel("Student_Depression_Dataset.xlsx")
print("#The primary goal of this dataset is to predict and analyze depression levels among students...")
print("\n#Dataset Information:")
print(f"Attributes: {df_info.shape[1]}, Samples: {df_info.shape[0]}")
print(df_info.dtypes)
print(df_info['Depression'].value_counts())

# Statistical summary and plots
from sklearn.preprocessing import MinMaxScaler, LabelEncoder, OneHotEncoder

uploaded = files.upload()
data = pd.read_excel('/content/Student_Depression_Dataset.xlsx')

sample_data = data.sample(n=50)
print("Sample Data:")
print(sample_data)

numeric_data = data.select_dtypes(include=[np.number])
print("\nMean Values:")
print(numeric_data.mean())
print("\nStandard Deviation Values:")
print(numeric_data.std())

plt.figure(figsize=(8, 5))
sns.histplot(data['Depression'], bins=20, kde=True)
plt.title('Distribution of Depression Scores')
plt.xlabel('Depression Score')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 5))
sns.scatterplot(x=data['Sleep Duration'], y=data['Depression'], alpha=0.6)
plt.title('Depression Score vs Sleep Hours')
plt.xlabel('Sleep Hours')
plt.ylabel('Depression Score')
plt.grid(True)
plt.show()
```
